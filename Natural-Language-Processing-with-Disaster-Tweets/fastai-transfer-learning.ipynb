```python
import numpy as np
import pandas as pd
from fastai.text.all import *
```

```python
train_path = "../input/nlp-getting-started/train.csv"
test_path = "../input/nlp-getting-started/test.csv"
sub_path = "../input/nlp-getting-started/sample_submission.csv"
```

```python
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
sub_df = pd.read_csv(sub_path)
```

```python
train_df.drop(["id", "keyword", "location"], inplace=True, axis=1)
```

```python
train_df.head()
```

```python
dls = TextDataLoaders.from_df(train_df)
dls.show_batch(max_n=2)
```

```python
learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, metrics=accuracy)
```

```python
learn.fine_tune(4, 1e-2)
```

```python
test_df.drop(["id", "keyword", "location"], inplace=True, axis=1)
```

```python
test_df.head()
```

```python
learn.show_results()
```

```python
dls = TextDataLoaders.from_df(train_df, shuffle=False)
```

```python
test_df.head()
```

```python
test_list = test_df["text"][:]
```

```python
int(learn.predict("It was a disaster!")[0])
```

```python
pred = np.zeros(len(test_list))
```

```python
for i, twt in enumerate(test_list): 
    pred[i] = int(learn.predict(twt)[0])
```

```python
pred
```

```python
sub_df["target"] = pred
```

```python
sub_df.head()
```

```python
test_df.head()
```

```python
sub_df['target'] = sub_df['target'].astype(int)
```

```python
sub_df.to_csv("submission.csv", index=False)
```
